```python
import pandas as pd
import pickle
from sklearn.metrics import f1_score, accuracy_score
import numpy as np
```

```python
langs = ['de', 'es', 'hu', 'bg', 'lv']
```

```python
pred_labels = {}
true_labels = {}

for i in langs:
    predictions = pickle.load(open(f'/content/drive/MyDrive/NLP/Project/MT/Labels/pred_labels_{i}.pkl', 'rb'))
    true = pickle.load(open(f'/content/drive/MyDrive/NLP/Project/MT/Labels/true_labels_{i}.pkl', 'rb'))

    pred_labels[i] = np.array(predictions).flatten()
    true_labels[i] = np.array(true).flatten()
```

```python
accuracy_scores = {}
micro_f1_scores = {}
macro_f1_scores = {}
weighted_f1_scores = {}

for i in langs:
    accuracy_scores[i] = accuracy_score(pred_labels[i], true_labels[i])
    micro_f1_scores[i] = f1_score(true_labels[i], pred_labels[i], average = 'micro')
    macro_f1_scores[i] = f1_score(true_labels[i], pred_labels[i], average = 'macro')
    weighted_f1_scores[i] = f1_score(true_labels[i], pred_labels[i], average = 'weighted')
```

```python
metrics = pd.DataFrame([accuracy_scores, micro_f1_scores, weighted_f1_scores, macro_f1_scores], index = ['Accuracy', 'Micro-f1', 'Weighted-f1', 'Macro-f1'])
metrics
```

```python
metrics.to_csv('/content/drive/MyDrive/NLP/Project/MT/metrics.csv')
```
